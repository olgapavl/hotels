```python
def to_lower_under_score(name):
    name = name.lower().replace(' ', '_')
    return(name)
df = df.rename(columns = to_lower_under_score)
```

```python
a = df[df['is_canceled'] == 0]
a['country'].value_counts()
```

```python
#На сколько ночей в среднем бронируют отели разных типов?
b = df.groupby('hotel').agg({'stays_total_nights': 'mean'}).round(2)
b
```

```python
#Иногда тип номера, полученного клиентом (assigned_room_type), отличается от изначально забронированного (reserved_room_type). 
#Такое может произойти, например, по причине овербукинга. 
#Сколько подобных наблюдений встретилось в датасете?
```

```python
df[df['assigned_room_type'] != df['reserved_room_type']].count()
```

```python
# На какой месяц чаще всего успешно оформляли бронь в 2016?
df[(df['arrival_date_year'] == 2017)&(df['is_canceled'] ==0)].groupby('arrival_date_month').count().sort_values('is_canceled',ascending=False)
```

```python
#Сгруппируйте данные по годам
#и проверьте, на какой месяц бронирования отеля типа City Hotel отменялись чаще всего в каждый из периодов
df[df['hotel'] =='City Hotel'].groupby(['arrival_date_year','arrival_date_month']).agg({'is_canceled': 'sum'})
```

```python
#Создайте колонку total_kids, объединив children и babies. 
df['total_kids'] = df['children'] + df['babies']
```

```python
df[df['total_kids'] != 0]
```

```python
#Отели какого типа в среднем пользуются большей популярностью у клиентов с детьми?
import numpy as np
df.groupby('hotel').agg({'total_kids': 'mean'})
```

```python
#Создайте переменную has_kids, которая принимает значение True, если клиент при бронировании указал хотя бы одного ребенка
#(total_kids), и False – в противном случае. 

df['has_kids'] = df.total_kids > 0 #создаем столбец по условию.
```

```python
#запрос по условию (нужно только количество строк,а результат запроса выдает кол-во строк и столбцов => shape[0]!!!)
df.query('is_canceled == 1 & has_kids == False').shape[0] / df.query('has_kids == False').shape[0] * 100
```

```python
df.query('is_canceled == 1 & has_kids == True').shape[0] / df.query('has_kids == True').shape[0] * 100
```
